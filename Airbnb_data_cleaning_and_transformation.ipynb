# Airbnb Open Data - Data Cleaning & Transformation

This notebook performs a full data cleaning on the Airbnb Open Data dataset.  
The steps covered are:

1. **Load Data** - Import the raw CSV
2. **Explore** - Inspect columns, dtypes, and missing values
3. **Drop Redundant Columns** - Remove low-quality or irrelevant features
4. **Rename Columns** - Standardise to UPPER_CASE
5. **Remove Duplicates** - Detect and drop exact duplicate rows
6. **Handle Missing Values** - Drop high-NaN columns and remaining NaN rows
7. **Clean Individual Columns** - Fix encoding issues and convert types
8. **Export** - Save the cleaned dataset as a CSV

## 1. Import library

```python
import pandas as pd 
```

## 2. Load Data

```python
df = pd.read_csv(
    '/kaggle/input/datasets/arianazmoudeh/airbnbopendata/Airbnb_Open_Data.csv',
    low_memory=False)
df.head()
```

## 3. Explore the dataset

```python
# List all column names in the raw dataset
df.columns
```

```python
df.info() # Show dtypes, non-null counts
```

## 4. Drop Redundant Columns
Based on the data exploration above, we split the 26 columns into two groups:
- **columns_to_keep** - 20 relevant features for analysis
- **columns_to_drop** - 6 columns that are not useful

```python
# Define the columns we want to keep for analysis
columns_to_keep = ['id', 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']

# Drop the column we dont want (low quality or not relevant)
columns_to_drop = ['reviews per month','review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license']
```

```python
print(f"Columns to keep : {len(columns_to_keep)}")
print(f"Columns to drop : {len(columns_to_drop)}")
```

```python
# Preview the columns we are keeping
keep_df = df[columns_to_keep]
keep_df.head()
```

```python
# Preview the columns we are dropping
drop_df = df[columns_to_drop]
drop_df.head()
```

```python
# Drop the redundant columns from the main dataframe 
df.drop(columns=columns_to_drop, inplace=True)

# The remaining columns
print("Remaining columns:")
df.columns
```

## 5. Rename Columns

Standardise all column names to *UPPER_CASE* for consistency:
- Convert to uppercase
- Replace spaces with underscores

```python
# Generate new column names: uppercase and spaces replaced with underscores
new_col_names = [col.upper().replace(" ", '_') for col in df.columns]

print("New column names:")
new_col_names
```

```python
# Create a working copy of the dataframe and apply the new column names

new_df_col_names = df.copy()
new_df_col_names.columns = new_col_names
df_new=new_df_col_names

df_new.head()
```

## 6. Remove Duplicate Rows

```python
# Count how many rows are exact duplicates before removal
df_new.duplicated().value_counts()
```

```python
# Drop all duplicate rows

df_new.drop_duplicates(inplace = True)
```

```python
# Confirm that no duplicates remain
print(f"Remaining duplicate rows: {df_new.duplicated().sum()}")
```

## 7. Handle Missing Values

**Strategy:**
- LAST_REVIEW has 15,893 missing values (15.5% of rows) - drop the entire column
- For remaining columns, drop any rows that still contain NaN values

```python
# count of missing values per column
df_new.isna().sum()
```

```python
df_new.drop(columns=["LAST_REVIEW"], inplace= True) #drop the last_review column
```

```python
# Drop all remaining rows that contain any NaN values
df_new.dropna(inplace=True)

print(f"Rows remaining after NaN removal: {len(df_new):,}")
```

```python
df_new.isna().sum() # verify for no NaN values 
```

## 8. Clean individual columns

- Convert boolean True/False values to 1/0 integers
- clean PRICE and SERVICE_FEE (Both columns has strings with currency symbols (eg.$) and change dtype from string(object) to int.

```python
df_new['INSTANT_BOOKABLE'] = df_new['INSTANT_BOOKABLE'].apply(lambda x: 1 if x == True else 0)

df_new['INSTANT_BOOKABLE'].value_counts()
```

```python
df_new['PRICE'].head()
```

```python
# Remove the "$" currency sign and "," thousands separator from PRICE
df_new['PRICE'] = df_new['PRICE'].str.replace("$", "", regex=False).str.replace(",", "", regex=False)

# Remove the same from SERVICE_FEE
df_new['SERVICE_FEE'] = df_new['SERVICE_FEE'].str.replace("$", "", regex=False).str.replace(",", "", regex=False)

df_new[['PRICE', 'SERVICE_FEE']].head()
```

```python
# PRICE and SERVICE_FEE from string (object) to integer dtype
df_new['PRICE'] = df_new['PRICE'].str.strip().astype(int)
df_new['SERVICE_FEE'] = df_new['SERVICE_FEE'].str.strip().astype(int)

print(df_new[['PRICE', 'SERVICE_FEE']].dtypes)
```

## 9. Index reset and final check

```python
# Reset the index after dropping duplicates and NaN rows
df_new.reset_index(drop=True, inplace=True)

# Preview the fully cleaned dataframe
df_new.head()
```

```python
# Final summary of the cleaned dataset
print(f"Shape       : {df_new.shape}")
print(f"Columns     : {list(df_new.columns)}")
print()
df_new.info()
```

## 10. Export cleaned data

```python
# Save the cleaned dataframe to CSV
df_new.to_csv('cleaned_airbnb_data.csv', index=False)

print("Cleaned dataset saved to: cleaned_airbnb_data.csv")
```
